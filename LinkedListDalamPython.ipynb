```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class LinkedList:
    def __init__(self):
        self.headval = None

    # Menampilkan linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # Menambah node di awal
    def AddBegining(self, newdata):
        NewNode = Node(newdata)
        NewNode.nextval = self.headval
        self.headval = NewNode

    # HAPUS NODE DI AKHIR
    def RemoveEnd(self):
        # Cek apakah linked list kosong
        if self.headval is None:
            print("Linked list kosong, tidak ada yang dihapus")
            return

        # Kasus 1: Hanya ada 1 node
        if self.headval.nextval is None:
            removed = self.headval
            self.headval = None
            print(f"Node '{removed.dataval}' (satu-satunya node) berhasil dihapus")
            return

        # Kasus 2: Ada lebih dari 1 node
        second_last = self.headval
        # Cari node sebelum node terakhir
        while(second_last.nextval.nextval):
            second_last = second_last.nextval

        # Simpan node terakhir untuk ditampilkan
        removed = second_last.nextval
        # Hapus node terakhir
        second_last.nextval = None
        print(f"Node akhir '{removed.dataval}' berhasil dihapus")

# Membuat linked list
Li = LinkedList()

# Menambahkan node
Li.AddBegining("Jumat")
Li.AddBegining("Kamis")
Li.AddBegining("Rabu")
Li.AddBegining("Selasa")
Li.AddBegining("Senin")

print("=" * 50)
print("Linked List AWAL:")
Li.listprint()
print("-" * 50)

# Hapus node di akhir
Li.RemoveEnd()

print("\nLinked List SETELAH hapus node di akhir:")
Li.listprint()
print("=" * 50)
```

```python
# buat node
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None  # Pointer ke node sebelumnya

class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    # Menampilkan double linked list dari depan ke belakang
    def listprint_forward(self):
        printval = self.headval
        print("Penelusuran maju (head -> tail):")
        while printval is not None:
            print(printval.dataval, end=" <-> ")
            printval = printval.nextval
        print("None")

    # Menampilkan double linked list dari belakang ke depan
    def listprint_backward(self):
        # Cari node terakhir dulu
        last = self.headval
        if last is None:
            print("Penelusuran mundur (tail -> head):")
            print("None")
            return
        while last.nextval is not None:
            last = last.nextval

        print("Penelusuran mundur (tail -> head):")
        while last is not None:
            print(last.dataval, end=" <-> ")
            last = last.prevval
        print("None")

    # Menambah node di akhir
    def AddEnd(self, newdata):
        NewNode = Node(newdata)

        # Jika linked list kosong
        if self.headval is None:
            self.headval = NewNode
            return

        # Cari node terakhir
        last = self.headval
        while last.nextval is not None:
            last = last.nextval

        # Hubungkan node baru
        last.nextval = NewNode
        NewNode.prevval = last

    # Penyisipan di TENGAH (setelah node dengan nilai tertentu)
    def InsertAfter(self, key, newdata):
        # Cari node dengan nilai key
        current = self.headval
        while current is not None:
            if current.dataval == key:
                break
            current = current.nextval

        # Jika key tidak ditemukan
        if current is None:
            print(f"Node dengan nilai '{key}' tidak ditemukan")
            return

        # Buat node baru
        NewNode = Node(newdata)

        # Hubungkan node baru dengan node setelahnya
        NewNode.nextval = current.nextval
        NewNode.prevval = current

        # Hubungkan node setelahnya dengan node baru (jika ada)
        if current.nextval is not None:
            current.nextval.prevval = NewNode

        # Hubungkan node saat ini dengan node baru
        current.nextval = NewNode

        print(f"Node '{newdata}' berhasil disisipkan setelah node '{key}'")

    # Hapus node dengan kata kunci
    def RemoveNode(self, key):
        current = self.headval

        # Cek apakah linked list kosong
        if current is None:
            print("Linked list kosong")
            return

        # Cari node dengan nilai key
        while current is not None:
            if current.dataval == key:
                break
            current = current.nextval

        # Jika key tidak ditemukan
        if current is None:
            print(f"Node dengan nilai '{key}' tidak ditemukan")
            return

        # Kasus 1: Node yang dihapus adalah HEAD
        if current.prevval is None:
            self.headval = current.nextval
            if self.headval is not None:
                self.headval.prevval = None
            print(f"Node HEAD '{key}' berhasil dihapus")

        # Kasus 2: Node yang dihapus adalah TAIL
        elif current.nextval is None:
            current.prevval.nextval = None
            print(f"Node TAIL '{key}' berhasil dihapus")

        # Kasus 3: Node yang dihapus ada di TENGAH
        else:
            current.prevval.nextval = current.nextval
            current.nextval.prevval = current.prevval
            print(f"Node TENGAH '{key}' berhasil dihapus")

        current = None
# soal
print("=" * 50)
print("Latihan 1: DOUBLE LINKED LIST")
print("=" * 50)

# Membuat double linked list dengan 3 node
dll = DoubleLinkedList()

# Menambahkan 3 node awal
dll.AddEnd("Node 1 (Awal)")
dll.AddEnd("Node 2 (Awal)")
dll.AddEnd("Node 3 (Awal)")

print("\nDouble Linked List AWAL (3 node):")
dll.listprint_forward()
dll.listprint_backward()
print("-" * 40)

# PENYISIPAN TENGAH
print("\nPENYISIPAN TENGAH:")
print("Menyisipkan 'Node Baru (Tengah)' setelah 'Node 2 (Awal)'")
dll.InsertAfter("Node 2 (Awal)", "Node Baru (Tengah)")

print("\nHasil setelah PENYISIPAN TENGAH:")
dll.listprint_forward()
dll.listprint_backward()
print("-" * 40)
```

```python
# Buat node
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None

class CircularLinkedList:
    def __init__(self):
        self.headval = None

    # Method to add a node to an empty list or at the end of a circular list
    def add_to_end(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval # Point to itself for circularity
        else:
            current = self.headval
            while current.nextval != self.headval:
                current = current.nextval
            current.nextval = NewNode
            NewNode.nextval = self.headval # New node points back to head

    # Method to print the circular linked list
    def listprint(self):
        if self.headval is None:
            print("Circular Linked List kosong")
            return

        printval = self.headval
        print("Penelusuran Circular Linked List (mulai dari head):")
        while True:
            print(printval.dataval, end=" -> ")
            printval = printval.nextval
            if printval == self.headval:
                break
        print(self.headval.dataval, "(kembali ke head)") # Indicate circularity


# IMPLEMENTASI SOAL 2
print("=" * 50)
print("Latihan 2: CIRCULAR LINKED LIST")
print("=" * 50)

# Membuat circular linked list dengan 3 node
cll = CircularLinkedList()

# Menambahkan 3 node awal
cll.add_to_end("Node A")
cll.add_to_end("Node B")
cll.add_to_end("Node C")

print("\nCircular Linked List AWAL (3 node):")
cll.listprint()
print("-" * 40)

# Contoh menambahkan node keempat
print("\nMenambahkan 'Node D' ke akhir list")
cll.add_to_end("Node D")

print("\nHasil setelah penambahan 'Node D':")
cll.listprint()
print("-" * 40)
```
