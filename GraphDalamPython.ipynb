1. Representasi Graf dengan Adjacency List


```python
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

def print_graph(graph):
    for node in graph:
        print(f"{node} -> {graph[node]}")

print_graph(graph)
```

2. Traversal Graf menggunakan BFS

```python
from collections import deque

def bfs(graph, start):
  visited = set()
  queue = deque([start])

  while queue:
    vertex = queue.popleft()
    if vertex not in visited:
      print(vertex, end=" ")
      visited.add(vertex)
      queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

bfs(graph, 'A')
```

3. Traversal Graf menggunakan DFS


```python
def dfs (graph, start, visited=None):
  if visited is None:
    visited = set ()

  if start not in visited:
    print (start, end=" ")
    visited.add(start)
    for neighbor in graph[start]:
      dfs (graph, neighbor, visited)

dfs (graph, 'A')
```

```python
def count_neighbors(graph):
    result = {}
    for node in graph:
        result[node] = len(graph[node])

    return result

base_graf = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A'],
    'D': ['B']
}
print(count_neighbors(base_graf))
```

1. Buatlah graf tak berarah baru dengan minimal 5 simpul dan 7 sisi, lalu tampilkan
dengan fungsi print_graph.
2. Implementasikan BFS dan DFS dari simpul awal pilihan Anda.
3. Modifikasi fungsi BFS agar mengembalikan list urutan kunjungan, bukan hanya
mencetak.
4. Buat fungsi find_path(graph, start, end) yang mengembalikan satu jalur dari start
ke end jika ada, menggunakan DFS.
5. Tambahkan fungsi is_connected(graph) untuk mengecek apakah semua simpul
dalam graf saling terhubung (gunakan kombinasi DFS + logika).

```python
from collections import deque

# ---------------------------------------------------------
# 1. Buatlah graf tak berarah baru dengan minimal 5 simpul dan 7 sisi
# ---------------------------------------------------------
def create_graph():
    # Representasi graf: dictionary { node: [tetangga] }
    # 5 Simpul: A, B, C, D, E
    # 7 Sisi: A-B, A-C, A-E, B-C, B-D, C-E, D-E
    graph = {
        'A': ['B', 'C', 'E'],
        'B': ['A', 'C', 'D'],
        'C': ['A', 'B', 'E'],
        'D': ['B', 'E'],
        'E': ['A', 'C', 'D']
    }
    return graph

def print_graph(graph):
    print("========Struktur Graf========")
    for node, neighbors in graph.items():
        print(f"{node} -> {neighbors}")
    print("=============================\n")

# ---------------------------------------------------------
# 2. Implementasikan BFS dan DFS (versi dasar tanpa return)
# ---------------------------------------------------------
def bfs_basic(graph, start):
    print(f"BFS Basic (Start: {start}):", end=" ")
    visited = set()           # Menandai simpul yang sudah dikunjungi
    queue = deque([start])    # Antrean untuk BFS
    visited.add(start)

    while queue:
        node = queue.popleft()
        print(node, end=" ")   # Cetak saat dikunjungi

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    print("\n")

def dfs_basic(graph, start, visited=None):
    if visited is None:
        visited = set()

    visited.add(start)
    print(start, end=" ")      # Cetak saat dikunjungi

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs_basic(graph, neighbor, visited)

# ---------------------------------------------------------
# 3. Modifikasi BFS untuk mengembalikan list urutan kunjungan
# ---------------------------------------------------------
def bfs_get_order(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)
    visit_order = []           # List untuk menyimpan urutan

    while queue:
        node = queue.popleft()
        visit_order.append(node) # Masukkan ke list, bukan dicetak

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return visit_order

# ---------------------------------------------------------
# 4. Buat fungsi find_path menggunakan DFS (Kembalikan satu jalur)
# ---------------------------------------------------------
def find_path(graph, start, end):
    def dfs_search(current, path):
        # Tambahkan node saat ini ke jalur
        path.append(current)

        # Basis: Jika sudah sampai di tujuan, kembalikan jalur saat ini
        if current == end:
            return path

        # Telusuri setiap tetangga
        for neighbor in graph[current]:
            if neighbor not in path:  # Hindari cycle
                result = dfs_search(neighbor, path)
                if result is not None:
                    return result

        # Jika tidak ada tetangga yang mencapai tujuan, backtrack (hapus node terakhir)
        path.pop()
        return None

    # Mulai pencarian
    final_path = dfs_search(start, [])
    return final_path

# ---------------------------------------------------------
# 5. Tambahkan fungsi is_connected (DFS + Logika)
# ---------------------------------------------------------
def is_connected(graph, start):
    if not graph:
        return True # Graf kosong dianggap terhubung

    visited = set()
    stack = [start]
    visited.add(start)

    # Lakukan DFS/Iterasi
    while stack:
        node = stack.pop()
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                stack.append(neighbor)

    # Jika jumlah node yang dikunjungi = total node dalam graf, maka terhubung
    return len(visited) == len(graph)

# ==================== MAIN PROGRAM ====================
if __name__ == "__main__":
    # 1. Buat Graf
    my_graph = create_graph()
    print_graph(my_graph)

    # 2. BFS & DFS Dasar
    print("2. Implementasi BFS & DFS:")
    bfs_basic(my_graph, 'A')
    print("DFS Basic (Start: A):", end=" ")
    dfs_basic(my_graph, 'A')
    print("\n\n")

    # 3. BFS Order List
    print("3. fungsi BFS Modifikasi:")
    order = bfs_get_order(my_graph, 'A')
    print(f"Urutan Kunjungan BFS: {order}\n")

    # 4. Find Path
    print("4. fungsi find_path (DFS):")
    start_node = 'A'
    end_node = 'D'
    path_result = find_path(my_graph, start_node, end_node)
    if path_result:
        print(f"Jalur dari {start_node} ke {end_node}: {' -> '.join(path_result)}")
    else:
        print(f"Tidak ada jalur dari {start_node} ke {end_node}")
    print()

    # 5. Is Connected
    print("5. Cek Keterhubungan:")
    if is_connected(my_graph, 'A'):
        print("Graf TERHUBUNG (Connected).")
    else:
        print("Graf TIDAK TERHUBUNG (Disconnected).")
```
