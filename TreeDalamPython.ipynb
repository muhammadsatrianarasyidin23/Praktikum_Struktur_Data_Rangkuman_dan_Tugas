1. Membuat Binary Tree Secara Manual

```python
class Node:
 def __init__(self, data):
  self.data = data
  self.left = None
  self.right = None

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

def inorder(node):
 if node:
  inorder(node.left)
  print(node.data, end=' ')
  inorder(node.right)

print("Hasil inorder traversal dari binary tree:")
inorder(root)
```

2. Membuat Binary Search Tree

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

class BinarySearchTree:
  def __init__(self):
    self.root = None
  def insert(self, root, data):
    if root is None:
      return Node(data)
    if data < root.data:
      root.left = self.insert(root.left, data)
    else:
      root.right = self.insert(root.right, data)
    return root

bst = BinarySearchTree()
root = None
data_list = [50, 30, 70, 20, 40, 60, 80]
for value in data_list:
  root = bst.insert(root, value)

print("BST berhasil dibuat dengan root:", root.data)
```

3. Traversal pada BST

```python
def inorder(node):
  if node:
    inorder(node.left)
    print(node.data, end=' ')
    inorder(node.right)

def preorder(node):
  if node:
    print(node.data, end=' ')
    preorder(node.left)
    preorder(node.right)

def postorder(node):
  if node:
    postorder(node.left)
    postorder(node.right)
    print(node.data, end=' ')

print("Inorder Traversal:")
inorder(root)
print("\nPreorder Traversal:")
preorder(root)
print("\nPostorder Traversal:")
postorder(root)
```

4. Pencarian Nilai dalam BST

```python
def search(node, key):
  if node is None or node.data == key:
    return node
  if key < node.data:
    return search(node.left, key)
  return search(node.right, key)

key = 60
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")

key = 25
result = search(root, key)
if result:
  print(f"{key} ditemukan dalam tree.")
else:
  print(f"{key} tidak ditemukan.")
```

Tugas 2: Program Binary Search Tree (BST)

```python
#Membuat class Node
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None

#Membuat class BST
class BinarySearchTree:
  def __init__(self):
    self.root = None
  #Fungsi insert data ke BST
  def insert(self, root, data):
    #Jika node kosong
    if root is None:
      return Node(data)
    #Jika data lebih kecil masuk kiri
    if data < root.data:
      root.left = self.insert(root.left, data)
    #Jika data lebih besar masuk kanan
    else:
      root.right = self.insert(root.right, data)
    return root

#Fungsi Inorder Traversal
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=' ')
        inorder(node.right)

#Fungsi Preorder Traversal
def preorder(node):
    if node:
        print(node.data, end=' ')
        preorder(node.left)
        preorder(node.right)

#Fungsi Postorder Traversal
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=' ')

# Fungsi Pencarian Data
def search(node, key):
    # Jika data ditemukan atau node kosong
    if node is None or node.data == key:
        return node
    # Cari ke subtree kiri
    if key < node.data:
        return search(node.left, key)
    # Cari ke subtree kanan
    return search(node.right, key)

#Data Daftar 7 Angka BST

#Tiga digit terakhir NPM: 018 -> 18
#Tanggal lahir: 03 dan 03
#ASCII huruf pertama dan kedua nama depan: M = 77, u = 117
#Dua angka: 23 dan 32

data_list = [18, 3, 3, 77, 117, 23, 32]

#membentuk BST
bst = BinarySearchTree()
root = None
for data in data_list:
    root = bst.insert(root, data)

#menampilkan Traversal
print("Inorder Traversal:")
inorder(root)
print("\n\nPreorder Traversal:")
preorder(root)
print("\n\nPostorder Traversal:")
postorder(root)

#pencarian Data
#Cari dua digit terakhir NPM = 18
key = 18
result = search(root, key)
if result:
    print(f"\n\nData {key} ditemukan dalam BST.")
else:
    print(f"\n\nData {key} tidak ditemukan.")


# Cari angka yang tidak ada dalam list
key = 327822
result = search(root, key)
if result:
    print(f"Data {key} ditemukan dalam BST.")
else:
    print(f"Data {key} tidak ditemukan.")
```
